```python
# Определить корни f(x) = 0.6x^3+5.5x^2+10x -5
from sympy import *
a = 0.6
b = 5.5
c = 10
d = -5

D = b*b - 4*a*c*d
print(f"D = {D}")

x = Symbol('x')
func=0.6*x**3 + 5.5*x**2 + 10*x - 5
y=solve(func)
x1=y[0]
x2=y[1]
x3=y[2]
print(f"Корни функции: x1 = ", x1, " x2 = ", x2, " x3 = ", x3)
```

```python
# Построить график
from sympy.plotting import plot
f = 0.6*x**3 + 5.5*x**2 + 10*x - 5
print(solve(f,x))
plot(f)
```

```python
# Найти интервалы, на которых функция возрастает
fd=diff(func)
print(solve(0<fd))
```

```python
# Найти интервалы, на которых функция убывает

print(solve(fd<0))
```

```python
# Вычислить вершину

corni=solve(fd)
top=corni[0]
x=top
y=0.6*x**3 + 5.5*x**2 + 10*x - 5
print(f"Координаты вершины: {top, y}")
```
